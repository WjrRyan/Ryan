```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import jieba

# 读取 CSV 文件
csv_file = '/Users/Ryan/Downloads/word9.csv'
df = pd.read_csv(csv_file)

# 特殊字符的列名，需要用反引号括起来
column_name = "JSONExtractString(request_body, 'keyword')"

# 提取列数据并生成文本
text = ' '.join(df[column_name].astype(str).tolist())

# 使用 jieba 对中文文本进行分词
words = ' '.join(jieba.cut(text))

# 创建词云对象，指定中文字体
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    font_path='/Library/Fonts/Songti.ttc'  # 替换为系统中存在的中文字体路径
).generate(words)

# 显示词云图像
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # 不显示坐标轴
plt.show()

# 保存词云图像
wordcloud.to_file('wordcloud0806.png')
```
