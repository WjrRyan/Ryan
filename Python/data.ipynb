```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 读取Excel文件
file_path = '/Users/Ryan/Desktop/data.xlsx'  # 替换为你的Excel文件路径
sheet_name = 'Sheet1'  # 替换为你的工作表名称
df = pd.read_excel(file_path, sheet_name=sheet_name)

# 假设你要统计的列名为 'data'
column_name = 'data'  # 替换为你的列名称

# 统计该列中不同值的数量
value_counts = df[column_name].value_counts()

# 设置绘图样式
sns.set(style="whitegrid")

# 创建柱状图
plt.figure(figsize=(10, 6))
sns.barplot(x=value_counts.index, y=value_counts.values, palette="viridis")

# 添加标题和标签
plt.title(f"Frequency Distribution of '{column_name}'", fontsize=16)
plt.xlabel(f"Unique Values in '{column_name}'", fontsize=14)
plt.ylabel("Frequency", fontsize=14)

# 显示图表
plt.xticks(rotation=45)  # 如果x轴标签较长，可以旋转以避免重叠
plt.tight_layout()  # 自动调整子图参数
plt.show()
```
