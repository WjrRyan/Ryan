```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 读取Excel文件
file_path = "/Users/Ryan/Downloads/地区系统行为分析_sr--2004031652.xlsx"
try:
    # 读取数据
    df = pd.read_excel(file_path)
    
    # 计算覆盖率
    df['覆盖率'] = (df['使用人数'] / df['总人数']) * 100
    
    # 数据分析
    print("覆盖率统计描述：")
    print(df['覆盖率'].describe())
    
    # 可视化
    plt.figure(figsize=(12, 8))
    
    # 覆盖率条形图
    plt.subplot(2, 2, 1)
    sns.barplot(data=df, x=df.index, y='覆盖率')
    plt.title('各行覆盖率分布')
    plt.xticks(rotation=45)
    
    # 覆盖率箱线图
    plt.subplot(2, 2, 2)
    sns.boxplot(y=df['覆盖率'])
    plt.title('覆盖率箱线图')
    
    # 覆盖率与使用人数的散点图
    plt.subplot(2, 2, 3)
    sns.scatterplot(data=df, x='使用人数', y='覆盖率')
    plt.title('使用人数与覆盖率关系')
    
    # 覆盖率分布直方图
    plt.subplot(2, 2, 4)
    sns.histplot(df['覆盖率'], bins=20)
    plt.title('覆盖率分布直方图')
    
    plt.tight_layout()
    plt.show()
    
    # 输出TOP5和BOTTOM5
    print("\n覆盖率最高的5行：")
    print(df.nlargest(5, '覆盖率')[['使用人数', '总人数', '覆盖率']])
    
    print("\n覆盖率最低的5行：")
    print(df.nsmallest(5, '覆盖率')[['使用人数', '总人数', '覆盖率']])

except KeyError as e:
    print(f"数据列不存在，请检查列名: {e}")
    print("\n实际的列名有：")
    print(df.columns.tolist())
except Exception as e:
    print(f"分析过程中发生错误：{str(e)}")
```

```python
import pandas as pd
import pygwalker as pyg

# 读取Excel文件
file_path = "/Users/Ryan/Downloads/地区系统行为分析_sr--2004031652.xlsx"
try:
    # 读取数据
    df = pd.read_excel(file_path)
    
    # 显示前几行数据
    print("数据预览：")
    print(df.head())
    
    # 显示基本信息
    print("\n数据基本信息：")
    print(df.info())
    
    # 创建新的列 'Usage_Rate'，等于 '使用人数' / '总人数'
    if '使用人数' in df.columns and '总人数' in df.columns:
        df['Usage_Rate'] = df['使用人数'] / df['总人数']
    else:
        print("未找到'使用人数'或'总人数'列，请检查列名称是否正确")

    # 使用 Pygwalker 进行数据分析
    pyg.walk(df)
    
except FileNotFoundError:
    print("文件未找到，请确认文件路径是否正确")
except Exception as e:
    print(f"读取文件时发生错误：{str(e)}")
```
