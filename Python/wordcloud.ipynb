```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import jieba
from collections import Counter
import random

# 添加停用词
stop_words = set(['的', '了', '和', '与', '或', '在', '是', '我们', '可以', '将', '有', '等', '来', '你', '会', 'The', 'to', 'and', 'in', 'of', 'a', 'the', 'how', 'with'])

# 自定义颜色函数
def color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']  # 设置配色方案
    return random.choice(colors)

# 分词并过滤停用词
words = jieba.cut(text)
words = [word for word in words if word.strip() and word not in stop_words]

# 统计词频
word_freq = Counter(words)

# 生成词云
wordcloud = WordCloud(
    width=1600,
    height=900,
    background_color='white',
    font_path='/System/Library/Fonts/PingFang.ttc',
    max_words=100,  # 减少词数以避免过于拥挤
    max_font_size=150,
    min_font_size=15,
    random_state=42,
    color_func=color_func,  # 使用自定义颜色
    prefer_horizontal=0.7,  # 70%的词水平显示
    collocations=False,
    margin=10,  # 添加边距
).generate_from_frequencies(word_freq)

# 显示词云
plt.figure(figsize=(20, 12), facecolor='white')
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()

# 打印最常见的20个词及其频率
print("\n最常见的词语：")
for word, freq in word_freq.most_common(20):
    print(f"{word}: {freq}")
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import jieba
from collections import Counter
import random
import numpy as np
from PIL import Image

# 将文本按类别明确分组
categories = {
    '技术类': """
    AIGC实践 PPT可视化 UX设计 骨骼动画 网页动效 
    视频制作 低代码开发 图表美化 音频处理 
    数字人技术 流程图制作 GIF动图 前端开发
    """,
    
    '管理与领导力': """
    团队管理 Leadership 新业务模式 TL角色转换
    核心价值观 管理能力 领导力提升 团队效能
    """,
    
    '客户服务': """
    服务体验 客户沟通 客诉处理 差评处理
    服务设计 高品质服务 酒店礼遇 应急救援
    客户满意度 服务优化
    """,
    
    '个人发展': """
    MBTI指南 效能提升 专业能力 时间管理
    个人成长 社交技巧 沟通能力 职业发展
    """,
    
    '产品与设计': """
    交互设计 服务设计 产品设计 竞品分析
    用户体验 产品创新 设计方法论 产品规划
    """
}

# 为每个类别分配固定的颜色
category_colors = {
    '技术类': '#2E86C1',      # 蓝色
    '管理与领导力': '#E74C3C', # 红色
    '客户服务': '#27AE60',    # 绿色
    '个人发展': '#F39C12',    # 橙色
    '产品与设计': '#8E44AD'   # 紫色
}

# 生成完整的文本，同时为每个词分配其类别的颜色
text_with_weights = {}
for category, text in categories.items():
    words = jieba.cut(text)
    for word in words:
        if len(word.strip()) > 1:
            # 为同一类别的词赋予相似的权重，使其在视觉上更加聚合
            text_with_weights[word] = {'text': word, 'weight': random.uniform(10, 20), 'color': category_colors[category]}

# 自定义颜色函数
def color_func(word, font_size, position, orientation, **kwargs):
    return text_with_weights[word]['color']

# 生成词云
wordcloud = WordCloud(
    width=1600,
    height=900,
    background_color='white',
    font_path='/System/Library/Fonts/PingFang.ttc',
    max_words=100,
    max_font_size=160,
    min_font_size=20,
    random_state=42,
    color_func=color_func,
    prefer_horizontal=0.7,
    collocations=False,
    margin=10
).generate_from_frequencies({word: data['weight'] for word, data in text_with_weights.items()})

# 显示词云
plt.figure(figsize=(20, 12), facecolor='white')
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.tight_layout(pad=0)

# 添加图例
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', 
                            markerfacecolor=color, label=category, markersize=10)
                  for category, color in category_colors.items()]
plt.legend(handles=legend_elements, loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```
